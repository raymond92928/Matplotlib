## 例子六

```python
# 新增數據資料一
Years = [1990,1995,2000,2005,2010,2015,2020]
Macau_Population = [34.38,38.55,42.78,48.29,53.82,60.21,64.93]

# 新增數據資料二
deaths_rate = [1.2,1.7,1.8,2.2,2.5,2.7,2.9]
deaths_Population = [round(Macau_Population[i]*deaths_rate[i]/100,2) for i in range(len(Years))]
print(deaths_Population)

# 分別繪線，設置參數
plt.plot(Years,Macau_Population, '.r-')
plt.plot(Years,deaths_Population, '.b--')

# 設置標題和軸描述
plt.title("MACAU Population")
plt.ylabel("Population(10k)")
plt.xlabel("Population growth by year")

# 設置網格
plt.grid(True)

# 顯示圖像
plt.show()
```

## 例子十一

```python
# 生成圖像
plt.figure(figsize = (10, 6))

#要生成兩行兩列，這是第一個圖plt.subplot('行','列','編號
x = np.arange(0.0,2.0,0.1)
y = np.sin(x*np.pi)
plt.subplot(2,2,1)
plt.plot(x,y,'r-')
plt.ylabel('y1')

#兩行兩列,這是第二個圖
x = np.arange(0.0,4.0,0.1)
y = np.sin(x*np.pi)
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.ylabel('y2')

#兩行兩列,這是第三個圖
x = np.arange(0.0,6.0,0.1)
y = np.sin(x*np.pi)
plt.subplot(2,2,3)
plt.plot(x,y,'m--')

#兩行兩列,這是第四個圖
x = np.arange(0.0,8.0,0.1)
y = np.sin(x*np.pi)
plt.subplot(2,2,4)
plt.plot(x,y,'c-')

# 顯示圖像
plt.show()
```

## 例子十二

```python
# 生成圖像
plt.figure(figsize=(12, 8))

for i in range(4):
    # 生成 兩行兩列 圖像，設置第 i 個圖像
    ax = plt.subplot(2,2,1+i)
    # 設置子圖標題
    alpha = 0.98 / 4 * i + 0.01
    ax.set_title(f'alpha={alpha:.3f}')
    # 生成數組
    t1 = np.arange(0.0, 1.0, 0.01)
    # 每張子圖繪製四條線
    for n in [1, 2, 3, 4]:
        plt.plot(t1, t1 ** n, label = f'n={n}')
    # 設置圖案說明
    leg = plt.legend(loc='upper left', shadow=True)

# 保存圖像
plt.savefig('figsize.jpg')

# 顯示圖像
plt.show()
```
