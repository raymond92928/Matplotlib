## 例子一：某某班男女百分比

```python
import matplotlib.pyplot as plt

# 輸入數據資料
data = [21, 22]
labels = ['male', 'female']

# 生成圖形
plt.figure()
# 載入數據資料，設置格式參數
plt.pie(data, labels=labels, autopct='%.2f%%')
# 顯示為圓形
plt.axis('equal')

plt.show()
```

## 例子二：全班同學最喜愛節目的百分比

```python
import matplotlib.pyplot as plt

# 輸入數據資料
# data = [0.08,0.20,0.30,0.36,0.06]
data = [4,10,15,18,3]
# labels = ['新聞','體育','動畫','娛樂','戲曲']
labels = ['news','sport','animation','entertainment','drama']

# 生成圖形 : 10x100, 6x100
plt.figure(figsize=(10, 6))
# 載入數據資料，設置格式參數
explode = (0, 0, 0, 0.1, 0)  # 突出效果
plt.pie(data, labels=labels, autopct='%.2f%%', explode=explode, shadow=True, startangle=90) # 立體效果、旋轉
#plt.pie(data, labels=labels, autopct='%.2f%%', explode=explode, shadow=True) # 立體效果、旋轉
# 設置標題
plt.title("percentage of the class's favorite show")

# 顯示為圓形
plt.axis('equal')

plt.show()
```

## 例子三：全班同學最喜愛節目的百分比(進階)

```python
import matplotlib.pyplot as plt

# 輸入數據資料
# data = [0.08,0.20,0.30,0.36,0.06]
data = [4,10,15,18,3]
# labels = ['新聞','體育','動畫','娛樂','戲曲']
labels = ['news','sport','animation','entertainment','drama']

# 自定義函數設置圖形圖說明文字
def make_autopct(values):
    def my_autopct(pct):
        val = int(round(pct*sum(values)/100.0))
        #\n下一行
        #d 整數
        return f'{pct:.2f}%\n({val:d})'
    return my_autopct

# 生成圖形
plt.figure(figsize=(10, 6))
# 載入數據資料，設置格式參數

explode = (0, 0, 0, 0.1, 0)  # 突出效果
plt.pie(data, autopct=make_autopct(data), explode=explode, shadow=True, startangle=90) # 立體效果、旋轉
#plt.pie(data,autopct='%.2f%%',  explode=explode, shadow=True, startangle=90) # 立體效果、旋轉

# 設置標題
plt.title("percentage of the class's favorite show")

# 設置顏色說明
plt.legend(labels, title="show", loc="center right", bbox_to_anchor=(1.2,0.5))

# 顯示為圓形
plt.axis('equal')

plt.show()
```
