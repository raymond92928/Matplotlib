```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 例子一：Students enrolled in different courses

```python
# 新增數據資料
data = {'C':5, 'C++':20, 'Java':10, 'Python':35, 'PHP':12}
courses = list(data.keys())
values = list(data.values())
print(courses)
print(values)

# 生成圖像
plt.figure(figsize = (10, 6))

# 將圖像設置為棒形圖，載入數據資料，設置格式參數
plt.bar(courses, values, color = 'maroon', width = 0.4)

# 設置標題和軸描述
plt.title("Students enrolled in different courses")
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")

# 顯示圖像
plt.show()
```

## 例子二：statistics of student answers

```python
# 新增數據資料
labels = np.array(["D", "C", "B", "A"])
values = np.array([3, 14, 22, 4])

# 生成圖像
plt.figure(figsize = (10, 6))

# 將圖像設置為橫向棒形圖(horizontal bars)，載入數據資料，設置格式參數
plt.barh(labels, values)
plt.get_cmap("viridis")

# 設置標題和軸描述
plt.title("statistics of student answers")
plt.xlabel("number of respondents")
plt.ylabel("choices")

# 儲存圖像
plt.savefig('sample_plot.jpg')

# 顯示圖像
plt.show()
```

## 例子三：scores by classes and gender

```python
# 新增數據資料
Male = (90, 92, 83, 90, 85, 80)
Female = (96, 94, 89, 86, 79, 82)

# 生成圖像
plt.figure(figsize = (10, 6))

# 設置 x 軸
N = len(Male)
ind = np.arange(N)
print('ind:',ind)
width = 0.35
plt.xticks(ind + width / 2, ('S3A', 'S3B', 'S3C', 'S3D', 'S3E', 'S3F'))

# 將圖像設置為棒形圖，載入數據資料，設置格式參數
plt.bar(ind, Male, width, label='Male')
plt.bar(ind + width, Female, width, label='Female')

# 設置標題和軸描述
plt.xlabel('classes')
plt.ylabel('scores')
plt.title('the average score by classes and gender')

# 設置網格
plt.grid(color='gray', linestyle='--', linewidth=1, axis='y', alpha=0.5)

# 設置顏色說明
plt.legend(loc="upper right")

# 顯示圖像
plt.show()
```

## 例子四：number of 2020 Summer Olympics medal

```python
# 新增數據資料
countries = ['USA', 'Japan' , 'GB', 'China', 'ROC', 'Germany', 'Australia']
bronzes = np.array([33, 17, 22, 18, 23, 16, 22])
silvers = np.array([41, 14, 20, 32, 28, 11, 7])
golds = np.array([39, 27, 22, 38, 20, 10, 17])

# 設置 x 軸
N = len(countries)
ind = np.arange(N)
print('ind:',ind)

# 生成圖像
plt.figure(figsize = (10, 6))

# 將圖像設置為棒形圖，載入數據資料，設置格式參數
plt.bar(ind, golds, width=0.8, label='golds', color='gold', bottom=silvers+bronzes)
plt.bar(ind, silvers, width=0.8, label='silvers', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.8, label='bronzes', color='#CD853F')

# 設置 x 軸的刻度
plt.xticks(ind, countries)

# 設置標題和軸描述
plt.xlabel("Countries")
plt.ylabel("Medals")
plt.title("number of 2020 Summer Olympics medal")

# 設置顏色說明
plt.legend(loc="upper right")

# 顯示圖像
plt.show()
```

## 例子五：number of 2020 Summer Olympics medal
如果我們想查看相同長度的條形圖，但作為該國家獲得的總獎牌的比例，我們可以執行以下操作：


```python
# 新增數據資料
countries = ['USA', 'Japan' , 'GB', 'China', 'ROC', 'Germany', 'Australia']
bronzes = np.array([33, 17, 22, 18, 23, 16, 22])
silvers = np.array([41, 14, 20, 32, 28, 11, 7])
golds = np.array([39, 27, 22, 38, 20, 10, 17])

# 設置 x 軸
N = len(countries)
ind = np.arange(N)
print('ind:',ind)

# 計算比例
total = bronzes + silvers + golds
proportion_bronzes = (bronzes / total) * 100
proportion_silvers = (silvers / total) * 100
proportion_golds = (golds / total) * 100
print(proportion_bronzes)
print(proportion_silvers)
print(proportion_golds)
print(proportion_bronzes + proportion_silvers + proportion_golds)

# 生成圖像
plt.figure(figsize = (10, 6))

# 將圖像設置為棒形圖，載入數據資料，設置格式參數
plt.bar(ind, proportion_golds, width=0.8, label='golds', color='gold', bottom=proportion_bronzes+proportion_silvers)
plt.bar(ind, proportion_silvers, width=0.8, label='silvers', color='silver', bottom=proportion_bronzes)
plt.bar(ind, proportion_bronzes, width=0.8, label='bronzes', color='#CD853F')

# 設置 x 軸的刻度
plt.xticks(ind, countries)

# 設置標題和軸描述
plt.xlabel("Countries")
plt.ylabel("Medals")
plt.title("2020 Summer Olympics medal by proportion")

# 顯示圖像
plt.show()
```
